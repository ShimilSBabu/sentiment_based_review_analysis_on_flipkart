# Importing Required Libraries

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch
import requests
from bs4 import BeautifulSoup
import re
from tqdm import tqdm
from time import sleep
```

# Setting the model and tokenizer

```python
tokenizer = AutoTokenizer.from_pretrained('nlptown/bert-base-multilingual-uncased-sentiment')
model = AutoModelForSequenceClassification.from_pretrained('nlptown/bert-base-multilingual-uncased-sentiment')
```

# Functions

## Function to read input details from user

```python
def input_details():

  # URL of reviews' page
  url = input("Enter URL : ")

  # Number of review pages to be checked
  review_page_count = int(input("Enter number of review pages to check : "))

  # Minimal length of review to be considered
  minimal_review_length = int(input("Enter minimum number of letters in a valid review : "))

  # Enter 1 if reviews needs to be displayed
  display_reviews = int(input("Enter 1 if reviews needs to be displayed. Otherwise enter 0 : "))

  return url, review_page_count, minimal_review_length, display_reviews
```

## Function to perform entiment analysis

```python
def sentiment_analisis(review):
  tokens = tokenizer.encode(review, return_tensors='pt')
  analisis = model(tokens)
  analisis.logits # unnormalized final scores of the model
  rating = int(torch.argmax(analisis.logits))+1
  return rating
```

# Main

```python
# Reading the input details from the user
URL, review_page_count, minimal_review_length, display_reviews = input_details()


sentiment_aggregate = 0
review_count = 0

# iterating over multiple pages of reviews
for page_num in tqdm(range(1,review_page_count)):
    page_url = URL + f'&page={page_num}'
    r = requests.get(page_url)    
    soup = BeautifulSoup(r.content, 'html.parser')
    cols = soup.find_all('div',attrs={'class':'col _2wzgFH K0kLPL'})
    sleep(1)
    for col in cols:
      row = col.find_all('div',attrs={'class':'row'})
      review = row[1].find('div').text

      if display_reviews:
        print(f'review {review_count} : ', review)

      if not review:
        break

      # print(review)
      if len(review) > minimal_review_length:
        sentiment_rating = sentiment_analisis(review)
        sentiment_aggregate += sentiment_rating

        if display_reviews:
          print('sentiment_rating for review {review_count} : ', sentiment_rating, '\n'*3)
        
        review_count += 1


if review_count:
  mean_rating = sentiment_aggregate/review_count
  print("\n"*8, 'THE INFERENCES \nThe avaerage rating of this product based on sentiment analysis of customer reviews is : ', mean_rating)

print(review_count, ' customer reviews was processed for reaching this inference.')
```
